# PERFORM DATA CLEANING BY APPLYING EVERYTHING WE LEARNED SO FAR!

```
import string
string.punctuation
```

```
import nltk # Natural Language tool kit
nltk.download('stopwords')
```

```
# You have to download stopwords Package to execute this command
from nltk.corpus import stopwords
stopwords.words('english')
```

```
# Let's define a pipeline to clean up all the messages 
# The pipeline performs the following: (1) remove punctuation, (2) remove stopwords

def review_cleaning(review):
  review_punc_removed = [char for char in review if char not in string.punctuation]
  review_punc_removed_join = "".join(review_punc_removed)
  review_punc_removed_join_clean = [word for word in review_punc_removed_join.split() if word.lower() not in stopwords.words('english')]
  return review_punc_removed_join_clean
```

```
# Let's test the newly added function
reviews_df_clean = reviews_df['verified_reviews'].apply(review_cleaning)
```

```
print(reviews_df_clean[3])
```

```
reviews_df_clean
```

```
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(analyzer=review_cleaning)
rereviews_tokenization = vectorizer.fit_transform(reviews_df['verified_reviews'])
```

```
print(vectorizer.get_feature_names_out())
```

```
reviews_df.drop(['verified_reviews'], axis=1, inplace=True)
reviews = pd.DataFrame(rereviews_tokenization.toarray())
```

```
# Now let's concatenate them together
reviews_df = pd.concat([reviews_df, reviews], axis=1)
```

# TRAIN A NAIVE BAYES CLASSIFIER MODEL

```
X.shape
```

```
y.shape
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```
X_train.shape
```

```
X_test.shape
```

```
from sklearn.naive_bayes import MultinomialNB
NB_classifier = MultinomialNB()
NB_classifier.fit(X_train, y_train)
```

```
y_pred = NB_classifier.predict(X_test)
y_pred
```

# ASSESS TRAINED MODEL PERFORMANCE  

```
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```
print('Accuracy {} %'.format( 100 * accuracy_score(y_pred, y_test)))
```

```
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True)
```

```
print(classification_report(y_test, y_pred))
```

# ASSIGNMENT - TRAIN AND EVALUATE A LOGISTIC REGRESSION CLASSIFIER

```
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```
y_pred = model.predict(X_test)
y_pred
```

```
from sklearn.metrics import accuracy_score
print('Accuracy {} %'.format( 100 * accuracy_score(y_test, y_pred)))
```

```
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True)
```

```
print(classification_report(y_test, y_pred))
```

# Dealing with unbalanced data using SMOTE strategy

```
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X_train, y_train)
```

```
# do the same process after creat Synthetic samples by appling SMOTE strategy
NB_classifier_res = MultinomialNB()
NB_classifier_res.fit(X_res, y_res)
```

```
y_pred_res = NB_classifier_res.predict(X_test)
y_pred_res
```

```
print('Accuracy {} %'.format( 100 * accuracy_score(y_test, y_pred_res)))
```

```
cm = confusion_matrix(y_test, y_pred_res)
sns.heatmap(cm, annot=True)
```

```
print(classification_report(y_test, y_pred_res))
```

```
model_res = LogisticRegression()
model_res.fit(X_res, y_res)
```

```
y_pred_res = model_res.predict(X_test)
y_pred_res
```

```
print('Accuracy {} %'.format( 100 * accuracy_score(y_test, y_pred_res)))
```

```
cm = confusion_matrix(y_test, y_pred_res)
sns.heatmap(cm, annot=True)
```

```
print(classification_report(y_test, y_pred_res))
```
